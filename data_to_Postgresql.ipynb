```python
import pandas as pd 
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('Data\Pozos_Colombia.csv')
df
```

```python
print(pd.io.sql.get_schema(df, name='colombia_wells'))
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/Wells_Colombia')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name='colombia_wells', con=engine))
```

```python
df.head(n=0).to_sql(name='colombia_wells', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='colombia_wells', con=engine, if_exists='append')
```
